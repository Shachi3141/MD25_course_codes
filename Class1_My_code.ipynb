### First code

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
y=np.sin(x)
z=np.cos(x)

plt.plot(x,y,label='sin(x)')
plt.plot(x,z,label='cos(x)')
plt.title('PLOT SIN AND cos')
plt.xlabel('X-------->')
plt.ylabel('Y-------->')
plt.grid()
plt.show()

```
